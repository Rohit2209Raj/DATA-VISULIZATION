```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as py
import seaborn as sns
```

Categorical Data Plots

Now let's discuss using seaborn to plot categorical data!

There are a few main plot types for this:

boxplot

violinplot

stripplot

swarmplot

barplot

countplot



```python
df=sns.load_dataset('tips')
df
```

CountPlot---->


```python
sns.countplot(df['sex'])
```

```python
sns.countplot(x=df['sex'])
```

```python
sns.countplot(x=df['sex'],hue=df['smoker'])
```



---


**Barplot-->**






---
x pe qualitative and y pe quantative



```python
sns.barplot(x=df['sex'],y=df['total_bill'])
# by default it compares with the mean;
```

```python
sns.barplot(x=df['sex'],y=df['total_bill'],estimator=np.sum)
```



---


**BoxPlot**

---



```python
sns.boxplot(x='tip',y='day',data=df,palette='rainbow')
```



---
