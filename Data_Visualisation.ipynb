```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Matplotlib Basics
```

```python
x=np.linspace(0,5,11)
x
y=x**2
y
```

```python
plt.plot(x,y,label="x vs x**2")



plt.title("x vs x**2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
# Subplot in Matplotlib
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("x vs x**2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.subplot(2,2,3)
plt.plot(x**2,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.subplot(2,2,4)
plt.plot(x,y**(1/2))
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.subplot(2,2,2)
plt.plot(y,x)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
#Object oriented way
```

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig=plt.figure(figsize=(30,10),dpi=100)
# size tuppe (width,height)
# dpi default 100
axis1=fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2=fig.add_axes([0.2,0.2,0.4,0.4])
axis2.plot(x,y)
```
